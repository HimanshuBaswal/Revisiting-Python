```python
# Create a scatter chart that showed the number of bike shares by temperature.
Figure, Axes = plt.subplots(figsize = (40,20))

c = np.random.randint(1,10,size = len(bsl))
s = np.random.randint(10,220,size = len(bsl))

Axes.scatter(bsl["Count"], bsl['Temperature'], s=s, c=c)
```

```python
Figure.savefig('Scatter_bike_share.png')
```

```python
# Create a pie chart that showed the distribution of bike shares by season.
labels =['Spring','Summer','Autumn','Winter']
counts = spring['Count'].sum(), summer['Count'].sum(), autumn['Count'].sum(), winter['Count'].sum()
```

```python
Figure_2, Axes_2 = plt.subplots(figsize = (14,8))

Axes_2.pie(counts, labels = labels, autopct = '%1.1f%%', explode = [0.1,0,0.1,0])
Axes_2.legend()
plt.show()
```
