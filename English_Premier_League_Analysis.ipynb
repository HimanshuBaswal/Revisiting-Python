### At new dataframe: Add a new column named “Penalty success rate”, and this column calculates and shows us the team’s penalty success rate.

```python
top_teams.info
```

```python
top_teams['Penalty_S_R'] = ((top_teams.penalty_save / top_teams.penalty_conceded) * 100).astype(int)
```

```python
top_teams
```

### Find the most successful team on penalty saving at new data frame.

```python
top_teams['Penalty_S_R'].max()
```

```python
top_teams[top_teams['Penalty_S_R'].max() == top_teams['Penalty_S_R']]
```

### At new dataframe: Create a bar chart that showed the distribution of teams by points.

```python
import matplotlib.pyplot as plt
```

```python
labels = ["mun","mci","che","liv","ars","tot"]
```

```python
plt.bar(top_teams['season'],top_teams['wins'])
plt.ylabel("Seasons")
plt.xlabel("Win Count")
plt.xticks = (labels)
plt.show()
```

### At new dataframe: Create a pie chart that showed the distribution of teams by goal attempts.

```python
labels = ["mun","mci","che","liv","ars","tot"]
```

```python
plt.pie(top_teams['total_scoring_att'], labels = labels, autopct = "%1.1f%%", explode = [0.1,0,0.1,0,0.1,0])
plt.legend(title ="Teams",bbox_to_anchor =(1, 0, 0.5, 1))
plt.show()
```
